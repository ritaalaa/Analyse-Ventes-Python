```python
import csv

def saisir_ventes():
    ventes = []
    n = int(input("Combien de ventes ? "))

    for i in range(n):
        v = float(input(f"Vente {i+1} : "))
        ventes.append(v)

    return ventes

def chiffre_affaire(ventes):
    return sum(ventes)

def analyse_avancee(ventes):
    if len(ventes) == 0:
        print("Aucune vente")
        return None

    moyenne = sum(ventes) / len(ventes)

    print("\n--- Résultats ---")
    print("Chiffre d'affaire :", chiffre_affaire(ventes))
    print("Moyenne :", moyenne)
    print("Max :", max(ventes))
    print("Min :", min(ventes))

    # Classement
    print("\n--- Classement des ventes ---")
    ventes_triees = sorted(ventes, reverse=True)
    for i, v in enumerate(ventes_triees):
        print(f"{i+1} - {v}")

    # Top performances ⭐
    print("\n--- Top performances ---")
    for v in ventes:
        if v >= moyenne * 1.2:
            print(v, "→ EXCELLENTE vente ⭐")

    # Alertes ⚠️
    print("\n--- Alertes ---")
    for v in ventes:
        if v < moyenne * 0.5:
            print(v, "→ Vente très faible ⚠️")

    return moyenne


# 🔹 CSV export
def enregistrer_csv(ventes, fichier="ventes.csv"):
    with open(fichier, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["prix"])
        for v in ventes:
            writer.writerow([v])


# 🔹 TXT report
def ecrire_rapport(ventes, moyenne, fichier="rapport.txt"):
    contenu = f"""===== RAPPORT DES VENTES =====

Nombre de ventes : {len(ventes)}
Chiffre d'affaire : {sum(ventes):.2f}
Moyenne : {moyenne:.2f}
Max : {max(ventes):.2f}
Min : {min(ventes):.2f}

==============================
"""

    with open(fichier, "w") as f:
        f.write(contenu)


# Programme principal
ventes = saisir_ventes()
moyenne = analyse_avancee(ventes)

if moyenne is not None:
    enregistrer_csv(ventes)
    ecrire_rapport(ventes, moyenne)
    print("\nFichiers générés : ventes.csv et rapport.txt")
```
